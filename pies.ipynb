```python
'''
Pies
CS 5001
Fall 2019
Homework 1

Test Plan:
Test Case #1
    crust: 1, meat: 1, corn: 1, potatoes: 1
    pies: 0
    Leftovers: 1 crust, 1 meat, 1 corn, 1 potato
    
Test Case #2
    crust: 5, meat: 10, corn: 3, potatoes: 6
    pies: 1
    Leftovers: 4 crust, 8 meat, 2 corn, 0 potatoes
    
Test Case #3
    crust: 3, meat: 5, corn: 2, potatoes: 25
    pies: 2
    Leftovers: 1 crust, 1 meat, 0 corn, 13 potatoes
    
Test Case #4
    crust: 10, meat: 14, corn: 6, potatoes: 24
    pies: 4
    Leftovers: 6 crust, 6 meat, 2 corn, 0 potatoes
    
Test Case #5
    crust: 10, meat: 10, corn: 10, potatoes: 20
    pies: 3
    Leftovers: 7 crust, 4 meat, 7 corn, 2 potatoes
'''


import math 

#establishing the constant variables needed to make a pie
need_crust = 1
need_meat = 2
need_corn = 1
need_potato = 6

def main():
    #prompting user to input the amount of each ingredient they have
    crust_answer = int(input('How many crusts do you have? '))
    meat_answer = int(input('How many packs of meat do you have? '))
    corn_answer = int(input('How many packs of corn do you have? '))
    potato_answer = int(input('How many potatoes do you have? '))

    #establishing the number of pies that could be made from each ingredient 
    crust = crust_answer / need_crust
    meat = meat_answer / need_meat
    corn = corn_answer / need_corn
    potato = potato_answer / need_potato

    #calculating the maximum number of pies that can be made by incorporating the amount of each ingredient into the minimum function
    max_pies = int(min(crust, meat, corn, potato))

    crust_used = crust_answer - max_pies

    #using the print function to inform the user how many pies can be made and what is leftover
    print ('Kiki made', max_pies, 'shepards pie(s) for the NSKS.\nYou and your friends donate the leftover ingrediants to the student program.')
    print ('Leftover:')
    #calculating leftovers my subtracting the amount of each ingredient used by the total quantity provided
    print (crust_answer - max_pies * need_crust, 'Crust(s)')
    print (potato_answer - max_pies * need_potato, 'Potoatoe(s)')
    print (meat_answer - max_pies * need_meat, 'Meat')
    print (corn_answer - max_pies * need_corn, 'Corn')

  
main()
```
