```python
import math
'''
Walk for a Cause
CS5001
Fall 2019
Homework 1

Goals: Experiential learning with variables, calculations, calling
external functions and problem decomposition. 

Test Plan:
Test Case #1
    3 km walk, 2 hr, $1/mile 
    1.86 miles, 0.93mph, pace - 64min & 31sec, earn $1.86
    
Test Case #2
    5 km walk, 1 hr, $2/mile 
    3.11 miles, 3.11mph, pace - 19min & 18sec, earn $3.11
    
Test Case #3
    0.3 km walk, 0.5 hr, $3.5/mile 
    0.43 miles, 0.87mph, pace - 68min & 58sec, earn $1.52
    
Test Case #4
    10 km walk, 3 hr, $5/mile 
    6.22 miles, 2.07mph, pace - 28min & 59sec, earn $31.1
    
Test Case #5
    2 km walk, 1 hr, $1.8/mile 
    1.24 miles, 1.24mph, pace - 48min & 23sec, earn $2.23
    
'''


def main():
    #prompting user to input how many km they walked and establishing that value with the float function
    user_km_answer = input('How many kilometers did you walk? ')
    km_walked = float(user_km_answer)

    #prompting user to input how long they walked and establishing that value with the float function
    user_time_answer = input('How many hours was your walk? ')
    hours_walked = float(user_time_answer)

    #prompting user to input how much money was pledged and establishing that value with the float function
    user_money_answer = input('How much is your sponsor paying, per mile? ')
    sponsor_money = float(user_money_answer)

    #calculating miles walked through conversion
    km_in_mile = 1.61
    miles_walked = km_walked / km_in_mile
    print ('Your miles walked is:', round(miles_walked, 2))

    #calculating the miles per hour
    mph = miles_walked / hours_walked 
    print ('Your miles per hour (MPH) is:', round(mph, 2)) 

    min_in_hour = 60
    sec_in_min = 60

    #calculating average pace
    pace_hours = hours_walked / miles_walked
    total_seconds = pace_hours * min_in_hour * sec_in_min
    sec_per_mile = float(total_seconds) / float(miles_walked)

    #converting the time from hours to minutes and seconds
    min_per_mile = int(sec_per_mile / sec_in_min)
    remainder = sec_per_mile - (min_per_mile * sec_in_min)
    print ('Your average pace is:', min_per_mile, 'minutes and', round(remainder, 2), 'seconds per mile')

    #calculating total earnings based on sponser money per mile and miles walked
    earnings = miles_walked * sponsor_money 
    print ('You earned $', round(earnings, 2), ' for the orphanage! Thanks!', sep = '')


main()
```
