```python
'''
Anna Daccache
CS5001
Fall 2019
11/12/19
HW 6
Balanced Parentheses

Test Plan

Practice Test 1:
phrase = [[[]]], balanced

Practice Test 2:
phrase = [{{()}}], balanced

Practice Test 3:
phrase = [{(})], not balanced

Practice Test 4:
phrase = ()(()(), not balanced

Practice Test 5:
phrase = {[({[()]})]}, balanced
'''

def Balanced_Parentheses(phrase):
    
    s = []
    x = ''
    
    for i in range(len(phrase)):
        if phrase[i] == '(' or phrase[i] == '[' or phrase[i] == '{':
            s.append(phrase[i]);
             
        # stack cannot be empty at this point
        # need a closing bracket
        if (len(s) == 0): 
            return False;  
  
        if phrase[i] == ')':
            # store the element  
            x = s.pop()
            # last opened should be the first closed
            # otherwise not balanced  
            if (x == '{' or x == '['): 
                return False;  
  
        elif phrase[i] == '}':  
            # store the element
            x = s.pop() 
            # last opened should be the first closed
            # otherwise not balanced  
            if (x == '(' or x == '['): 
                return False;  
          
        elif x == ']':  
            # store the element 
            x = s.pop() 
            # last opened should be the first closed
            # otherwise not balanced
            if (x =='(' or x == '{'): 
                return False;  
  
    # check the empty stack  
    if len(s): 
        return True;
    else : 
        return False;


if __name__ == '__main__':  
  
    phrase = input('Write a series of parantheses: ')  
  
    if Balanced_Parentheses(phrase) == True: 
        print('Balanced')
    else:
        print('Not Balanced')
```
